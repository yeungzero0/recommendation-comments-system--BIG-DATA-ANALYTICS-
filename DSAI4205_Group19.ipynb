# Task4 visualization function

```python
# Task4 function
def clean_text(text):
    # Basic cleaning: lower, ensure string, remove newlines
    if pd.isna(text):
        return ""
    return str(text).lower().replace("\n", " ").strip()

def visualize_model_performance(y_test, y_pred, y_prob, X_test, titleName):
    # Convert to safe pandas/numpy types
    y_test = pd.Series(y_test).reset_index(drop=True).astype(int)
    y_pred = pd.Series(y_pred).reset_index(drop=True).astype(int)

    # Normalize y_prob to a 1d float array (try to handle common shapes)
    y_prob = np.array(y_prob)
    if y_prob.ndim == 2:
        # If shape (n,2) assume column 1 is positive class
        if y_prob.shape[1] == 2:
            y_prob = y_prob[:, 1]
        elif y_prob.shape[1] == 1:
            y_prob = y_prob.ravel()
        else:
            # fallback: try to ravel
            y_prob = y_prob.ravel()
    else:
        y_prob = y_prob.ravel()
    # Force numeric and finite
    y_prob = pd.Series(y_prob).astype(float)
    if not np.isfinite(y_prob).all():
        print("Warning: y_prob contains non-finite values; replacing with 0.0")
        y_prob = y_prob.fillna(0.0).replace([np.inf, -np.inf], 0.0)
    y_prob = y_prob.reset_index(drop=True).to_numpy()

    # Basic checks for ROC/PR plotting
    can_plot_roc = True
    can_plot_pr = True
    unique_labels = np.unique(y_test)
    if not set(unique_labels).issubset({0, 1}):
        print(f"Warning: y_test contains non-binary labels: {unique_labels}. Casting to 0/1.")
        y_test = (y_test != 0).astype(int)

    if len(np.unique(y_test)) < 2:
        print("ROC/PR unavailable: y_test contains only one class.")
        can_plot_roc = can_plot_pr = False

    if np.nanmin(y_prob) < 0 or np.nanmax(y_prob) > 1:
        print(f"Warning: y_prob range is [{np.nanmin(y_prob)}, {np.nanmax(y_prob)}]. Clipping to [0,1].")
        y_prob = np.clip(y_prob, 0.0, 1.0)

    if np.allclose(y_prob, y_prob[0]):
        print("ROC/PR unavailable: predicted probabilities are constant.")
        can_plot_roc = can_plot_pr = False

    # Prepare masks / groups
    pos_mask = y_test == 1
    neg_mask = y_test == 0
    mis_mask = (y_test != y_pred)

    # Clean text for word clouds / counts
    X_test = pd.Series(X_test).reset_index(drop=True).fillna("").astype(str).map(clean_text)

    pos_corpus = X_test[pos_mask].tolist()
    neg_corpus = X_test[neg_mask].tolist()
    mis_corpus = X_test[mis_mask].tolist()

    # Create figure 2x3
    fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(20, 10))
    axes = axes.flatten()
    plt.subplots_adjust(hspace=0.35, wspace=0.25)
    fig.suptitle(titleName, fontsize=18, y=0.95)

    # Confusion matrix
    ax = axes[0]
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax)
    ax.set_title("7. Confusion Matrix")
    ax.set_xlabel("Predicted")
    ax.set_ylabel("True")

    # ROC curve
    ax = axes[1]
    if can_plot_roc:
        try:
            fpr, tpr, _ = roc_curve(y_test, y_prob)
            roc_auc = auc(fpr, tpr)
            ax.plot(fpr, tpr, label=f"AUC = {roc_auc:.3f}", color='darkorange')
            ax.plot([0, 1], [0, 1], linestyle='--', color='gray')
            ax.set_xlim([0.0, 1.0])
            ax.set_ylim([0.0, 1.05])
            ax.set_xlabel('False Positive Rate')
            ax.set_ylabel('True Positive Rate')
            ax.set_title('8. ROC Curve')
            ax.legend(loc="lower right")
        except Exception as e:
            ax.text(0.5, 0.5, 'ROC unavailable', ha='center')
            ax.axis('off')
            print("ROC plotting error:", e)
    else:
        ax.text(0.5, 0.5, 'ROC unavailable', ha='center')
        ax.axis('off')

    # Precision-Recall curve
    ax = axes[2]
    if can_plot_pr:
        try:
            precision, recall, _ = precision_recall_curve(y_test, y_prob)
            pr_auc = auc(recall, precision)
            ax.plot(recall, precision, label=f"AUC = {pr_auc:.3f}", color='purple')
            ax.set_xlabel('Recall')
            ax.set_ylabel('Precision')
            ax.set_title('9. Precision-Recall Curve')
            ax.legend(loc='lower left')
        except Exception as e:
            ax.text(0.5, 0.5, 'PR unavailable', ha='center')
            ax.axis('off')
            print("PR plotting error:", e)
    else:
        ax.text(0.5, 0.5, 'PR unavailable', ha='center')
        ax.axis('off')

    # Histogram of prediction probabilities
    ax = axes[3]
    sns.histplot(y_prob, bins=30, kde=False, ax=ax, color='steelblue')
    ax.set_title("10. Histogram of predicted probabilities (positive class)")
    ax.set_xlabel("Predicted probability")
    ax.set_ylabel("Count")

    # True vs Predicted class distribution
    ax = axes[4]
    true_counts = y_test.value_counts().sort_index()
    pred_counts = pd.Series(y_pred).value_counts().sort_index()
    idx = ['0', '1']
    true_vals = [true_counts.get(0, 0), true_counts.get(1, 0)]
    pred_vals = [pred_counts.get(0, 0), pred_counts.get(1, 0)]
    x = np.arange(len(idx))
    width = 0.35
    ax.bar(x - width/2, true_vals, width, label='True', color='tab:blue')
    ax.bar(x + width/2, pred_vals, width, label='Predicted', color='tab:orange')
    ax.set_xticks(x)
    ax.set_xticklabels(idx)
    ax.set_title("11. True vs Predicted class distribution")
    ax.set_xlabel("Class")
    ax.set_ylabel("Count")
    ax.legend()
    
    # Top-10 word bar chart for misclassified samples
    ax = axes[5]
    if len(mis_corpus) > 0:
        try:
            # Build word counts
            from collections import Counter
            import string

            # Use STOPWORDS from WordCloud as in the original code
            stopwords_set = set(STOPWORDS) if 'STOPWORDS' in globals() else set()

            all_words = []
            for doc in mis_corpus:
                # simple tokenization: split on whitespace after clean_text already applied
                tokens = [tok.strip(string.punctuation).lower() for tok in doc.split() if tok.strip(string.punctuation)]
                # filter stopwords and empties and short tokens
                tokens = [t for t in tokens if t and t not in stopwords_set and len(t) > 1]
                all_words.extend(tokens)

            if len(all_words) == 0:
                ax.text(0.5, 0.5, 'No eligible words in misclassified texts', ha='center')
                ax.axis('off')
            else:
                counter = Counter(all_words)
                top_n = 10
                most_common = counter.most_common(top_n)
                words, counts = zip(*most_common)
                # Plot horizontal bar chart
                y_pos = np.arange(len(words))[::-1]  # reverse for descending top->bottom
                ax.barh(y_pos, counts, color='tab:green')
                ax.set_yticks(y_pos)
                ax.set_yticklabels(words)
                ax.set_xlabel('Count')
                ax.set_title(f"12. Top {len(words)} words in Misclassified (n={len(mis_corpus)})")
                # annotate counts on bars
                for i, v in enumerate(counts[::-1]):
                    ax.text(v + max(counts)*0.01, i, str(v), va='center')
        except Exception as e:
            ax.text(0.5, 0.5, 'Error generating word counts', ha='center')
            ax.axis('off')
            print("Word count / bar chart error:", e)
    else:
        ax.text(0.5, 0.5, 'No misclassified samples', ha='center')
        ax.axis('off')

    plt.tight_layout(rect=[0, 0.0, 1, 0.95])
    plt.show()



def visualize_dataset(df):
    dfc = df.copy()
    
    # compute numeric stats for review length
    dfc['reviewText'] = dfc['reviewText'].astype(str).fillna("")
    dfc['review_length'] = dfc['reviewText'].apply(len)

    # Prepare figure with 2 rows x 4 cols
    fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(24, 10))
    axes = axes.flatten()
    plt.subplots_adjust(hspace=0.4, wspace=0.3)
    fig.suptitle('Dataset Visualization', fontsize=18, y=0.98)

    # Prepare positive / negative text subsets
    pos_texts = dfc[dfc['label'] == 1]['reviewText'].astype(str)
    neg_texts = dfc[dfc['label'] == 0]['reviewText'].astype(str)
    
    # Helper to plot top-k terms from a CountVectorizer result as bar chart
    def plot_top_terms(ax, texts, ngram_range=(1,1), top_k=10, color='tab:blue', title='Top terms'):
        ax.cla()
        texts = texts[texts.str.strip() != ""]
        if texts.shape[0] == 0:
            ax.text(0.5, 0.5, 'No text available', ha='center')
            ax.axis('off')
            return
        try:
            cv = CountVectorizer(stop_words='english', ngram_range=ngram_range, max_features=5000)
            X = cv.fit_transform(texts)
            sums = np.asarray(X.sum(axis=0)).ravel()
            if sums.sum() == 0:
                ax.text(0.5, 0.5, 'No tokens after stopword filtering', ha='center')
                ax.axis('off')
                return
            terms = cv.get_feature_names_out()
            order = sums.argsort()[::-1][:top_k]
            top_terms = [terms[i] for i in order]
            top_counts = sums[order]
            # horizontal bar for readability
            y_pos = np.arange(len(top_terms))[::-1]
            ax.barh(y_pos, top_counts, color=color)
            ax.set_yticks(y_pos)
            ax.set_yticklabels(top_terms)
            ax.set_xlabel('Count')
            ax.set_title(title)
            # annotate counts
            max_count = top_counts.max() if len(top_counts)>0 else 0
            for i, v in enumerate(top_counts[::-1]):
                ax.text(v + max_count*0.01, i, str(int(v)), va='center', fontsize=9)
        except Exception as e:
            ax.text(0.5, 0.5, 'Error generating terms', ha='center')
            ax.axis('off')
            print(f"Error in plot_top_terms (ngram_range={ngram_range}):", e)

    
    # unigrams (1) -> top unigrams for positive
    ax = axes[0]
    plot_top_terms(ax, pos_texts, ngram_range=(1,1), top_k=10, color='tab:green', title='Unigrams - Positive (label=1)')

    # bigrams (1) -> top bigrams for positive
    ax = axes[1]
    plot_top_terms(ax, pos_texts, ngram_range=(2,2), top_k=10, color='tab:green', title='Bigrams - Positive (label=1)')

    # unigrams (0) -> top unigrams for negative
    ax = axes[2]
    plot_top_terms(ax, neg_texts, ngram_range=(1,1), top_k=10, color='tab:red', title='Unigrams - Negative (label=0)')
    
    # bigrams (0) -> top bigrams for negative
    ax = axes[3]
    plot_top_terms(ax, neg_texts, ngram_range=(2,2), top_k=10, color='tab:red', title='Bigrams - Negative (label=0)')

    #Original classification distribution (1..5)
    ax = axes[4]
    try:
        order = sorted(dfc['classification'].dropna().unique())
        sns.countplot(x='classification', data=dfc, palette='muted', order=order, ax=ax)
        ax.set_title('Original Star Rating Distribution (1..5)')
        ax.set_xlabel('Star rating')
        ax.set_ylabel('Count')
    except Exception as e:
        ax.text(0.5, 0.5, 'Error plotting distribution', ha='center')
        ax.axis('off')
        print("Plot error (original distribution):", e)

    # Binary class distribution (small separate figure)
    ax = axes[5]
    sns.countplot(x='label', data=dfc, palette='Set2', ax=ax)
    ax.set_title('Binary Label Distribution (0: Negative, 1: Positive)')
    
    
    # Top 10 Positive reviews word cloud -> bar chart
    ax = axes[6]
    plot_top_terms(ax, pos_texts, ngram_range=(1,1), top_k=10, color='tab:green', title='Top 10 terms - Positive (label=1)')

    #Top 10 Negative reviews word cloud -> bar chart
    ax = axes[7]
    plot_top_terms(ax, neg_texts, ngram_range=(1,1), top_k=10, color='tab:red', title='Top 10 terms - Negative (label=0)')
    
    plt.tight_layout(rect=[0, 0.0, 1, 0.96])
    plt.show()

    # # Boxplot review length by binary label 
    # plt.figure(figsize=(6,4))
    # sns.boxplot(x='label', y='review_length', data=dfc, palette='Set3')
    # plt.title('Review length distribution by label')
    # plt.show()


```

# "Accuracy", "Precision", "Recall", "F1-score", "AUC-ROC"  visualization 

```python
    
def safe_metrics(y_true, y_pred, y_prob):
    return {
        'accuracy': accuracy_score(y_true, y_pred),
        'precision': precision_score(y_true, y_pred, zero_division=0),
        'recall': recall_score(y_true, y_pred, zero_division=0),
        'f1': f1_score(y_true, y_pred, zero_division=0),
        'auc': roc_auc_score(y_true, y_prob)
    }

def display_metrics(metrics, title):
    metrics_df = pd.DataFrame({
        "Metric": ["Accuracy", "Precision", "Recall", "F1-score", "AUC-ROC"],
        "Value": [metrics.get('accuracy'), metrics.get('precision'), metrics.get('recall'), metrics.get('f1'), metrics.get('auc')]
    })

    # Plot metrics
    plt.figure(figsize=(10, 6))
    plt.plot(metrics_df["Metric"], metrics_df["Value"], marker='o', linestyle='-', label=title)
    plt.xlabel("Metrics")
    plt.ylabel("Value")
    plt.title(f"{title} Graph")
    plt.legend()
    plt.grid()
    plt.show()
    print(title, metrics)
```

# visualization for top10 amazon_reviews

```python
def top10Result(X_test, y_test, star_test, ensemble_prob,
                misclassified_words=None,
                return_df=False,
                save=False):
     
    csv_path='./result.csv'
    demotion_for_star_threshold=2
    demotion_for_star_threshold=2
    star_demotion_factor=0.3
    misword_demotion_factor=0.75
    top_k=10
    
    # Align inputs as pandas Series
    Xs = pd.Series(X_test).reset_index(drop=True).astype(str).fillna("")
    stars = pd.Series(star_test).reset_index(drop=True)
    probs = pd.Series(np.array(ensemble_prob).ravel()).reset_index(drop=True)

    # Ensure probabilities numeric and in [0,1]; if passed labels 0/1 they'll be cast to floats
    probs = pd.to_numeric(probs, errors='coerce').fillna(0.0).clip(0.0, 1.0)

    # Basic alignment length check
    if not (len(Xs) == len(stars) == len(probs) and (y_test is None or len(y_test) == len(probs))):
        raise ValueError("Length mismatch between X_test, star_test, y_test (if provided), and ensemble_prob")

    # If misclassified_words not provided, compute from misclassified samples in this set:
    if misclassified_words is None:
        try:
            ys = pd.Series(y_test).reset_index(drop=True).astype(int)
            preds = (probs >= 0.5).astype(int)
            mis_mask = (ys != preds)
            mis_texts = Xs[mis_mask].astype(str).tolist()
        except Exception:
            mis_texts = []

        # Extract frequent words from mis_texts (simple tokenizer)
        stopwords_set = set(STOPWORDS) if 'STOPWORDS' in globals() else set()
        all_words = []
        for doc in mis_texts:
            tokens = [tok.strip(string.punctuation).lower() for tok in doc.split() if tok.strip(string.punctuation)]
            tokens = [t for t in tokens if t and t not in stopwords_set and len(t) > 1]
            all_words.extend(tokens)
        counter = Counter(all_words)
        misclassified_words = set([w for w, _ in counter.most_common(50)])  # top-50 suspicious words
    else:
        misclassified_words = set([w.lower() for w in misclassified_words])

    # Build result DataFrame
    df_final = pd.DataFrame({
        'review': Xs,
        'star': stars,
        'ensemble_prob': probs
    }).reset_index(drop=True)

    # Predictions and flags
    df_final['pred_label'] = (df_final['ensemble_prob'] >= 0.5).astype(int)
    df_final['is_misleading'] = (df_final['pred_label'] == 1) & (df_final['star'] <= demotion_for_star_threshold)

    # Detect if review contains any misclassified word
    def contains_misword(text):
        tokens = [tok.strip(string.punctuation).lower() for tok in text.split() if tok.strip(string.punctuation)]
        for t in tokens:
            if t in misclassified_words:
                return True
        return False

    df_final['contains_misclassified_word'] = df_final['review'].apply(contains_misword)

    # Base trust score = ensemble probability
    df_final['trust_score'] = df_final['ensemble_prob'].astype(float)

    # Apply demotions multiplicatively
    df_final.loc[df_final['is_misleading'], 'trust_score'] *= float(star_demotion_factor)
    df_final.loc[df_final['contains_misclassified_word'], 'trust_score'] *= float(misword_demotion_factor)

    # Add matched miswords list (string)
    def list_matched_miswords(text):
        tokens = [tok.strip(string.punctuation).lower() for tok in text.split() if tok.strip(string.punctuation)]
        matched = sorted(set([t for t in tokens if t in misclassified_words]))
        return ",".join(matched)

    df_final['matched_misclassified_words'] = df_final['review'].apply(list_matched_miswords)

    # Save full DataFrame to CSV (overwrites if exists)
    if save:
        try:
            df_final.to_csv(csv_path, index=False, encoding='utf-8-sig')
            print(f"Saved full results to: {csv_path}")
        except Exception as e:
            print(f"Warning: could not save CSV to {csv_path}: {e}")

    # Sort by trust_score descending and take top_10
    top_reviews = df_final.sort_values(by='trust_score', ascending=False).head(top_k).reset_index(drop=True)

    # Display top_10
    print("\nTop {} Reviews to Promote (with demotions applied):".format(top_k))
    display(top_reviews[['review', 'star', 'ensemble_prob', 'trust_score', 'is_misleading',
                         'contains_misclassified_word', 'matched_misclassified_words']])

    if return_df:
        return top_reviews
    else:
        return None
```

```python
# Task2: RandomForest on BERT embeddings

#BERT
bert_model = SentenceTransformer('all-MiniLM-L6-v2')
X_train_bert = bert_model.encode(X_train.tolist(), show_progress_bar=True)
X_val_bert = bert_model.encode(X_val.tolist(), show_progress_bar=True)
X_test_bert = bert_model.encode(X_test.tolist(), show_progress_bar=True)


# RandomForest - Find best D
max_depth_values = [2,3,4,5,6,8,10]
best_auc = -1
best_d = None
nes = 100 #25,50,100,200, n=200(0.8437), n=100(0.8474), n=50(0.8443), n=25(0.8050)

for d in max_depth_values:
    rf_tmp = RandomForestClassifier(max_depth=d, n_estimators=nes, class_weight='balanced', random_state=SD)
    rf_tmp.fit(X_train_bert, y_train)
    val_probs = rf_tmp.predict_proba(X_val_bert)[:,1]
    val_auc = roc_auc_score(y_val, val_probs)
    print(f"depth={d}, val AUC={val_auc:.4f}")
    if val_auc > best_auc:
        best_auc = val_auc
        best_d = d

print("Selected depth on validation:", best_d, "with val AUC:", best_auc)

# Train RandomForest on BERT embeddings
rf = RandomForestClassifier(max_depth=best_d, n_estimators=nes, class_weight='balanced', random_state=SD)
rf.fit(X_train_bert, y_train)
y_pred_rf_bert = rf.predict(X_test_bert)
y_prob_rf_bert = rf.predict_proba(X_test_bert)[:,1]

# Evaluation 
metrics_rf_bert = safe_metrics(y_test, y_pred_rf_bert, y_prob_rf_bert)
display_metrics(metrics_rf_bert, "Task 2.1 - RF-BERT metrics")

#task4
visualize_model_performance(y_test, y_pred_rf_bert, y_prob_rf_bert, X_test, "Task 4 - RF on BERT [Task2.1]")

top10Result(X_test,y_test,star_test,y_pred_rf_bert)
```
